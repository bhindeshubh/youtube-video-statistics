### 12. Write a function to replace NA values with median for following column 'Trend_day_count', 'Tag_count', 'Trend_tag_count', 'comment_count', 'likes', 'dislike', 'views', 'subscriber'

```python
def ReplaceNumNull(data):
  data['Trend_day_count'] = data['Trend_day_count'].fillna(data['Trend_day_count'].median())
  data['Tag_count'] = data['Tag_count'].fillna(data['Tag_count'].median())
  data['Trend_tag_count'] = data['Trend_tag_count'].fillna(data['Trend_tag_count'].median())
  data['comment_count'] = data['comment_count'].fillna(data['comment_count'].median())
  data['likes'] = data['likes'].fillna(data['likes'].median())
  data['dislike'] = data['dislike'].fillna(data['dislike'].median())
  data['views'] = data['views'].fillna(data['views'].median())
  data['subscriber'] = data['subscriber'].fillna(data['subscriber'].median())
  return data.isnull().sum()
```

```python
df3 = df2.copy()
ReplaceNumNull(df3)
```

### 13. Write a function to treat ouliers also check the skewness of the data

```python
# Function to treat outliers
def treat_outliers(data, column):
    Q1 = data[column].quantile(0.25)  # First quartile
    Q3 = data[column].quantile(0.75)  # Third quartile
    IQR = Q3 - Q1  # Interquartile range
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    data[column] = np.where(data[column] < lower_bound, lower_bound, data[column])
    data[column] = np.where(data[column] > upper_bound, upper_bound, data[column])
```

```python
df4 = df3.copy()
for col in df4.select_dtypes('number').columns:
    treat_outliers(df4, col)
```

```python
plt.figure(figsize=(15, 20))

colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd',
          '#8c564b', '#e377c2', '#7f7f7f']

num_colors = len(colors)

c = 1
for i, col2 in enumerate(df4.select_dtypes('number').columns):
    plt.subplot(8, 2, c)

    train_color = colors[c % num_colors]

    sns.boxplot(data=df4[col2], color=train_color)

    plt.ylabel(col2)
    plt.title(f'Boxplot after outlier treatment of {col2}')

    c += 1

plt.tight_layout()
plt.show()
```
